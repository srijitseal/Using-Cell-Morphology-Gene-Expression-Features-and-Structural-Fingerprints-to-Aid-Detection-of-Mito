```python
import pandas as pd
```

```python
df= pd.read_csv("../../../../Ecker_et_al_Assays/Ecker_et_al_assays_Inchi_final.csv")
df= df.rename(columns={"InChICode_standardised" : "StdInChI"})
df
```

```python
pred_CP_GO= pd.read_csv("../Predicitions_for_CP_GO_Space_no_cell_deathv4.csv")
pred_CP_GO
```

```python
ext_test=pd.merge(df, pred_CP_GO, how='inner', on=['StdInChI'])
ext_test
```

```python
ext_test["CP_Correct?"] = (ext_test["y_pred_CP_Model"]==ext_test["Activity"]).astype(str)
ext_test["GO_Correct?"] = (ext_test["y_pred_GO_Model"]==ext_test["Activity"]).astype(str)
ext_test["Morgan_Correct?"] = (ext_test["y_pred_Morgan_Model"]==ext_test["Activity"]).astype(str)
ext_test["Early_Fusion_CP_GO_Model_Correct?"] = (ext_test["y_pred_Early_Fusion_CP_GO_Model"]==ext_test["Activity"]).astype(str)
ext_test["Early_Fusion_Morgan_CP_Correct?"] = (ext_test["y_pred_Early_Fusion_Morgan_CP"]==ext_test["Activity"]).astype(str)
ext_test["Early_Fusion_CP_GO_Morgan_Correct?"] = (ext_test["y_pred_Early_Fusion_CP_GO_Morgan_Model"]==ext_test["Activity"]).astype(str)
ext_test["Late_Stage_Fusion_CP_GO_Morgan_Correct?"] = (ext_test["y_pred_Late_Stage_Fusion_CP_GO_Morgan_Model"]==ext_test["Activity"]).astype(str)
ext_test
```

```python
crossval =pd.read_csv("../GO_CP_MitoOverlap_nocelldeath.csv")
crossval =crossval[crossval.PUBCHEM_ACTIVITY_OUTCOME != "Inconclusive"]
crossval= crossval.rename(columns={"PUBCHEM_ACTIVITY_OUTCOME" : "Activity"})

crossval = crossval[["StdInChI","Activity"]]
ext_test = ext_test[["StdInChI","Activity", "CP_Correct?", "GO_Correct?", "Morgan_Correct?",
                     "Early_Fusion_CP_GO_Model_Correct?","Early_Fusion_Morgan_CP_Correct?", "Early_Fusion_CP_GO_Morgan_Correct?", "Late_Stage_Fusion_CP_GO_Morgan_Correct?"]]
```

```python
crossval["Data"]= "CrossVal"
ext_test["Data"]= "ExtTest"
crossval
```

```python
ext_test
```

```python


crossval.to_csv("Crossval_compounds.csv", index=False)
ext_test.to_csv("ext_test_compounds.csv", index=False)
```

```python
all_data_chemical_space=pd.concat([crossval, ext_test])
all_data_chemical_space = all_data_chemical_space.reset_index(drop=True)
all_data_chemical_space
```

```python
from rdkit import Chem, DataStructs
from rdkit.Chem import AllChem
from rdkit.Chem import inchi
from rdkit import Chem

def get_standardize_smiles(value):
    try: return Chem.MolToSmiles(Chem.inchi.MolFromInchi(value))
    except: return "Cannot_do"
    
from pandarallel import pandarallel
pandarallel.initialize()

all_data_chemical_space['smiles_r']=all_data_chemical_space['StdInChI'].parallel_apply(get_standardize_smiles)  
```
