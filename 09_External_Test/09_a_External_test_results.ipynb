```python
import pandas as pd
```

```python
df= pd.read_csv("../../../../Ecker_et_al_Assays/Ecker_et_al_assays_Inchi_final.csv")
df= df.rename(columns={"InChICode_standardised" : "StdInChI"})
df
```

```python
#pred_CP_GO= pd.read_csv("Predicitions_for_CP_GO_Space_no_cell_death.csv")
pred_CP_GO= pd.read_csv("../Predicitions_for_CP_GO_Space_no_cell_deathv4.csv")
pred_CP_GO
```

```python
merged=pd.merge(df, pred_CP_GO, how='inner', on=['StdInChI'])
#merged.to_csv("blind.csv", index=False)
merged
```

```python
merged.Activity.value_counts()
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import average_precision_score
from sklearn.metrics import matthews_corrcoef,confusion_matrix, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
```

```python
from math import pi
```

```python
y_true= merged["Activity"].to_numpy()


y_pred_list=["y_pred_CP_Model", "y_pred_GO_Model", "y_pred_Morgan_Model", "y_pred_Early_Fusion_CP_GO_Model",
            "y_pred_Early_Fusion_Morgan_CP", "y_pred_Early_Fusion_CP_GO_Morgan_Model",
            "y_pred_Late_Stage_Fusion_CP_GO_Morgan_Model"]

y_prob_list= ["y_proba_CP_Model", "y_proba_GO_Model", "y_proba_Morgan_Model", "y_proba_Early_Fusion_CP_GO_Model",
            "y_proba_Early_Fusion_Morgan_CP", "y_proba_Early_Fusion_CP_GO_Morgan_Model",
            "y_proba_Late_Stage_Fusion_CP_GO_Morgan_Model"]
models = ["CP Model", "GO Model", "Morgan Model", "Early Fusion CP+GO Model",
            "Early Fusion Morgan+CP", "Early Fusion CP+GO+Morgan Model",
            "Late Stage Fusion CP+GO+Morgan Model" ]

list_of_lists = []



for (i,j,k) in zip(y_pred_list, y_prob_list, models):
    
    print(k)
    
    y_prob = merged[j].to_numpy()
    y_pred = merged[i].to_numpy()
    
    
    
    conf_matrix = confusion_matrix(y_true, y_pred)
    print(conf_matrix)
    print(classification_report(y_true, y_pred))
    ba= balanced_accuracy_score(y_true, y_pred)
    print('balanced_accuracy_score ',ba)
    mcc=matthews_corrcoef(y_true, y_pred)
    print('matthews_corrcoef ',mcc)
            # Sensitivity, hit rate, recall, or true positive rate 
    Specificity = conf_matrix[0,0]/(conf_matrix[0,0]+conf_matrix[0,1])
            # Specificity or true negative rate
    Sensitivity = conf_matrix[1,1]/(conf_matrix[1,0]+conf_matrix[1,1]) 
    
    f1= f1_score(y_true, y_pred, average='binary')
    precision = precision_score(y_true, y_pred, average='binary')
    
    print('F1 Toxic', f1)
    print('Precision', precision)
    print('Sensitivity', Sensitivity)
    print('Specificity', Specificity)
    AUC = roc_auc_score(y_true, y_prob)
    print('AUC-ROC ',AUC)  
    AUCPR = average_precision_score(y_true, y_prob, average='weighted')
    print('AUCPR ',AUCPR) 
            # calculate roc curves
    best_model_fpr, best_model_tpr, _ = roc_curve(y_true, y_prob)   
    plt.plot(best_model_fpr, best_model_tpr, marker='.', label='Our Model')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()
    plt.show()
    
    row=[ k, f1, precision, Sensitivity, Specificity, ba, mcc, AUC, AUCPR]
    list_of_lists.append(row)
```

```python
df_results = pd.DataFrame(list_of_lists, columns=["Type", "F1 (Toxic)" , "Precision (Toxic)", "Sensitivity", "Specificity", "BA", "MCC" , "AUC", "AUCPR"])
df_results
```

```python
df_results.to_csv("external_results.csv")
```

```python
def make_spider( row, title, color):

    # number of variable
    categories=list(df_results)[1:]
    N = len(categories)

    # What will be the angle of each axis in the plot? (we divide the plot / number of variable)
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]

    # Initialise the spider plot
    ax = plt.subplot(3,3,row+1, polar=True, )

    # If you want the first axis to be on top:
    ax.set_theta_offset(pi / 2)
    ax.set_theta_direction(-1)

    # Draw one axe per variable + add labels labels yet
    plt.xticks(angles[:-1], categories, color='grey', size=9)

    # Draw ylabels    
    ax.set_rlabel_position(0)
    plt.yticks([0.20,0.40,0.60,0.80], ["0.20","0.40","0.60","0.80"], color="grey", size=7)
    plt.ylim(0,1)

    # Ind1
    values=df_results.loc[row].drop('Type').values.flatten().tolist()
    values += values[:1]
    ax.plot(angles, values, color=color, linewidth=3, linestyle='solid')
    ax.fill(angles, values, color=color, alpha=0.9)

    # Add a title
    plt.title(title, size=11, color=color, y=1.1)

    
# ------- PART 2: Apply the function to all individuals
# initialize the figure
plt.figure(figsize=(20, 20), dpi=300)
 
# Create a color palette:
my_palette = plt.cm.get_cmap("Set2", len(df_results.index))
 
# Loop to plot
for row in range(0, len(df_results.index)):
    make_spider( row=row, title=df_results['Type'][row], color=my_palette(row))
```
