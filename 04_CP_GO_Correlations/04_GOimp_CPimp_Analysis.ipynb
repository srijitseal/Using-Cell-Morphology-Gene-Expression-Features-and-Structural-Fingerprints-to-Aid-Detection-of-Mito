```python

CP_imp_list_Analyse_list = CP_imp_list[1:]

GO_imp_list_Analyse_list =[
   
'activation of signaling protein activity involved in unfolded protein response (GO:0006987)',
'intrinsic apoptotic signaling pathway in response to endoplasmic reticulum stress (GO:0070059)',
'cellular response to unfolded protein (GO:0034620)',
'ER-nucleus signaling pathway (GO:0006984)',
'regulation of T cell apoptotic process (GO:0070232)',
"side of membrane (GO:0098552)",    
'positive regulation of endopeptidase activity (GO:0010950)',    
'regulation of oxidative stress-induced neuron death (GO:1903203)',    
'endoplasmic reticulum unfolded protein response (GO:0030968)',   
'mitotic DNA integrity checkpoint (GO:0044774)',   
'negative regulation of glucose import (GO:0046325)',    
'negative regulation of glucose transport (GO:0010829)',    
'repressing transcription factor binding (GO:0070491)',   
'response to endoplasmic reticulum stress (GO:0034976)']
```

```python
result2 = pd.concat([CP_imp_df[CP_imp_list_Analyse_list], GO_imp_df[GO_imp_list_Analyse_list]], axis=1).corr(method='pearson')
result2 = result2[CP_imp_list_Analyse_list].loc[GO_imp_list_Analyse_list]
result2 = result2.T
result2
```

```python
sns.clustermap(result2,figsize=(15, 40))
```

```python
from scipy.stats import pearsonr
result2pdf = pd.concat([CP_imp_df[CP_imp_list_Analyse_list], GO_imp_df[GO_imp_list_Analyse_list]], axis=1)
result2p = result2pdf.corr(method=lambda x, y: pearsonr(x, y)[1]) - np.eye(len(result2pdf.columns)) 
result2p = result2p[CP_imp_list_Analyse_list].loc[GO_imp_list_Analyse_list]
result2p = result2p.T
result2p = -np.log10(result2p)
result2p
```

```python
sns.clustermap(result2p,figsize=(15, 40))
```

```python
final_imp_data_pvalue=pd.DataFrame(columns= ['GO', 'pvalue', 'CP',])

for Feature in CP_imp_list_Analyse_list:

    trial_data=  result2p[result2p.index==Feature]
    trial_data = trial_data.T
    
    #trial_data[Feature]=abs(trial_data[Feature])
    trial_data = trial_data.sort_values(by=Feature, ascending=False)
    
    trial_data= trial_data.rename_axis("GO").reset_index()
    
    trial_data["CP"]= Feature 
    
    trial_data= trial_data.rename(columns={Feature: "pvalue"})
    
    frames = [final_imp_data_pvalue,trial_data]
    final_imp_data_pvalue = pd.concat(frames)
final_imp_data_pvalue ["Combination"] = final_imp_data_pvalue ["GO"] + " "+ final_imp_data_pvalue ["CP"]
final_imp_data_pvalue
```

```python
final_imp_data_statistic=pd.DataFrame(columns= ['GO', 'Corr', 'CP',])

for Feature in CP_imp_list_Analyse_list:

    trial_data=  result2[result2.index==Feature]
    trial_data = trial_data.T
    
    #trial_data[Feature]=abs(trial_data[Feature])
    trial_data = trial_data.sort_values(by=Feature, ascending=False)
    
    trial_data= trial_data.rename_axis("GO").reset_index()
    
    trial_data["CP"]= Feature 
    
    trial_data= trial_data.rename(columns={Feature: "Corr"})
    
    frames = [final_imp_data_statistic,trial_data]
    final_imp_data_statistic = pd.concat(frames)
    
final_imp_data_statistic ["Combination"] = final_imp_data_statistic ["GO"] + " "+ final_imp_data_statistic ["CP"]
final_imp_data_statistic =final_imp_data_statistic.drop(columns=["CP", "GO"], axis=1)
final_imp_data_statistic
```

```python
final_imp_data = pd.merge(final_imp_data_pvalue, final_imp_data_statistic, on=["Combination"])
final_imp_data
```
