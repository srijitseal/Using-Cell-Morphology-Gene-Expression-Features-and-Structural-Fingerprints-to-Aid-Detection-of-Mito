```python
import pandas as pd
```

```python
df = pd.read_csv("../Mitotox_all_datawarrior.csv")
df = df.drop(columns=['Common_name', 'iupac_name', 'smiles_r', 'InChICode_standardised'], axis=1)
df
```

```python
df=df[df["Cells_Number_Object_Number"]>-15]
df=df[df.PUBCHEM_ACTIVITY_OUTCOME!=3]
```

```python
df
```

```python
data=df
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.linear_model import LogisticRegression

from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from tqdm import tqdm
```

```python
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import classification_report
from sklearn.metrics import matthews_corrcoef
from sklearn.metrics import roc_auc_score
from sklearn.metrics import average_precision_score
from sklearn.metrics import f1_score

col_list=[]
TP_list=[]
FP_list=[]
TN_list=[]
FN_list=[]

balanced_accuracy_score_list=[]
precision_score_list=[]
mcc_list=[]
spe_list=[]
sen_list=[]
auc_list=[]
f1_list=[]
averqage_precision_list=[]

list_of_lists=[]

for column in tqdm(data.columns[1:]):
    
    df=data[[column, "PUBCHEM_ACTIVITY_OUTCOME"]]
    df=df[pd.notnull(df[column])]
    
    X = df[column].values.reshape(-1, 1)
    #print(X)
    y = df["PUBCHEM_ACTIVITY_OUTCOME"].values.reshape(-1, 1)
    #print(y)
    
    clf = DecisionTreeClassifier(random_state=42, max_depth=1, min_samples_split=10, min_samples_leaf=5, max_leaf_nodes=2)
    clf.fit(X, y)
  
    y_proba = clf.predict_proba(X)[:,1]
    y_true = y
    y_pred = clf.predict(X)

    conf_matrix = confusion_matrix(y_true, y_pred)
    TN=conf_matrix[0, 0]
    FP=conf_matrix[0, 1]
    FN=conf_matrix[1, 0]
    TP=conf_matrix[1, 1]
    
    TP_list.append(TP)
    FP_list.append(FP)
    TN_list.append(TN)
    FN_list.append(FN)
    
    col_list.append(column)  
    
    mcc=matthews_corrcoef(y_true, y_pred)
    mcc_list.append(mcc)
    
    auc= roc_auc_score(y_true, y_proba)
    auc_list.append(auc.round(2))
    
    #Average Precision is a single number used to summarise a Precision-Recall curve
    averqage_precision = average_precision_score(y_true, y_proba) 
    averqage_precision_list.append(averqage_precision.round(2))
    
    f1=f1_score(y_true, y_pred, average='binary')
    f1_list.append(f1.round(2))
    
    Specificity = TN/(TN+FP)
    spe_list.append(Specificity.round(2))
    Sensitivity = TP/(TP+FN)
    sen_list.append(Sensitivity.round(2))
    
    balanced_accuracy_score_list.append(balanced_accuracy_score(y_true, y_pred).round(2))
    precision_score_list.append(precision_score(y_true, y_pred, average='binary').round(2))

list_of_lists=[col_list, TP_list, FP_list, TN_list, FN_list, 
               precision_score_list, 
               balanced_accuracy_score_list,
               auc_list,
               averqage_precision_list,
               f1_list,
               mcc_list,
               sen_list,
               spe_list,]
data_ppv = pd.DataFrame(list_of_lists).transpose()
data_ppv.columns = ['Feature',"TP", "FP", "TN", "FN", 'PPV', 'BA', "AUC_ROC", "AUC_PR", "F1_Score", "MCC", "Sensitivity", "Specificity"]
```
