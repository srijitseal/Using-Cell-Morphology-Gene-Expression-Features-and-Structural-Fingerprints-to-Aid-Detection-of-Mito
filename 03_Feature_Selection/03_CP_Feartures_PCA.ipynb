```python
dw = pd.read_csv("PCA_ready_DW_542_CP110.csv")
dw
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
scalar.fit(dw[CP_features[1:]])
df_std = scalar.transform(dw[CP_features[1:]])
df_std = pd.DataFrame(df_std, columns= dw[CP_features[1:]].columns)
df_std["Activity"]=dw["PUBCHEM_ACTIVITY_OUTCOME"]
df_std
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
pca_all = pca.fit_transform(df_std[CP_features[1:]])

PC1= "PC1 (explained variance: " + str(np.round_(pca.explained_variance_ratio_[0]*100, decimals=1)) + "%)"
PC2= "PC2 (explained variance: " + str(np.round_(pca.explained_variance_ratio_[1]*100, decimals=1)) + "%)"
PC3= "PC3 (explained variance: " + str(np.round_(pca.explained_variance_ratio_[2]*100, decimals=1)) + "%)"
df_std.loc[:, PC1] = pca_all.T[0]
df_std.loc[:, PC2] = pca_all.T[1] 
df_std.loc[:, PC3] = pca_all.T[2] 

sns.jointplot(x=PC1, y=PC2, data=df_std)
g = sns.FacetGrid(df_std, col="Activity", hue="Activity")
g = (g.map(plt.scatter, PC1, PC2, edgecolor="w", s=10))
np.cumsum(pca.explained_variance_ratio_)
```
